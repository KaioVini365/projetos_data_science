## 1. Similaridade entre usuários via produto escalar

Dados dois vetores que representam preferências de usuários em três categorias diferentes, `user1 = [4, 3, 2]` e `user2 = [1, 5, 4]`, calcule a similaridade entre esses usuários usando o produto escalar.

```python
user1 = np.array([4, 3, 2])
user2 = np.array([1, 5, 4])

similaridade = np.dot(user1, user2)
print(similaridade)
```

## 2. Multiplicação escalar aplicada a dados reais

Dada uma série de preços de ações `prices = [120, 125, 130, 128, 135]`, ajuste os preços em 15% e calcule o novo preço das ações.

```python
prices = np.array([120, 125, 130, 128, 135])
taxa_reajuste = 0.15

novos_precos = prices + (prices * taxa_reajuste)  # vetorizado: sem loop sobre cada preço
print(novos_precos)
```

## 3. Otimização do produto escalar

Encontre o vetor unitário que maximiza o produto escalar com o vetor `v = [3, 4, 5]`.

**Solução analítica** — pela identidade cos(θ) = (u·v)/(‖u‖‖v‖), o produto escalar é máximo quando os vetores são paralelos e apontam na mesma direção (θ = 0°), o que dá `u = v/‖v‖` diretamente.

```python
v = np.array([3, 4, 5])
norma_v = np.linalg.norm(v)
u = v / norma_v  # vetor unitário na mesma direção de v

print(u)
print(np.linalg.norm(u))  # confirma que u é unitário (norma 1)
```

```python
# confirma que u·v == ||v||, que é o maior produto escalar possível com um vetor unitário
print(np.dot(u, v))
print(np.linalg.norm(v))
```

**Validação numérica** — checagem cruzada da solução analítica via otimização restrita com `scipy.optimize.minimize`.

```python
from scipy.optimize import minimize

def funcao_objetivo(x):
    return -np.dot(v, x)  # minimizar o negativo do produto escalar == maximizá-lo

restricao = {'type': 'eq', 'fun': lambda x: np.linalg.norm(x) - 1}  # x precisa ser vetor unitário

resultado = minimize(funcao_objetivo, np.ones(3), constraints=restricao)
vetor_otimizado = resultado.x
print(vetor_otimizado)
```

```python
escalar_maximo = np.dot(v, vetor_otimizado)
escalar_maximo
```

## 4. Produto vetorial e direção normal

Dados três pontos no espaço 3D `A = [1, 2, 3]`, `B = [4, 5, 6]` e `C = [7, 8, 9]`, encontre um vetor normal ao plano definido por esses três pontos.

```python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])
C = np.array([7, 8, 9])

AB = B - A
AC = C - A
vetor_normal = np.cross(AB, AC)
print(vetor_normal)  # [0, 0, 0]: os pontos são colineares, não definem um plano
```
